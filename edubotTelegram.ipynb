```python
import os
import google.generativeai as genai
from telegram import Update
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes
import asyncio

# Configura la clave de API de Google Generative AI y el token de Telegram
os.environ["GOOGLE_API_KEY"] = 'AIzaSyDWWmB9auEMsQXUaL47b2BzgulVwFhnX30'  # Reemplaza con tu clave de API real
TELEGRAM_API_TOKEN = '7629272567:AAF130qSnzuD-2Py5CdBNEHMGytnGFNm5DY'  # Token de Telegram

# Configura el modelo generativo
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
model = genai.GenerativeModel('gemini-1.5-flash-latest')
chat = model.start_chat(history=[])

# Función del chatbot que recibe mensajes y responde
async def chatbot(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    input_text = update.message.text
    response = chat.send_message(input_text)
    await update.message.reply_text(response.text)

# Inicia el bot de Telegram en modo asíncrono
async def main():
    # Crea la aplicación con el token
    application = Application.builder().token(TELEGRAM_API_TOKEN).build()

    # Maneja mensajes de texto con la función chatbot
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, chatbot))

    # Inicia el bot en modo asíncrono
    await application.initialize()
    await application.start()
    print("Bot en ejecución. Presiona Ctrl+C para detenerlo.")

    # Ejecuta indefinidamente
    await application.updater.start_polling()

# Ejecuta el bot en el bucle asíncrono del entorno
await main()
```
